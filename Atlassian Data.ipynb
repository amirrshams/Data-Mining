```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
```

```python
with open('atlassian data.csv', 'w') as file:
    file.write("titles; categories; nofinstalls \n")
```

```python
driver = webdriver.Chrome()
url = 'https://marketplace.atlassian.com/search'
driver.get(url)
```

```python
#while (True):
driver.find_element_by_xpath('//*[@id="amkt-frontend-content"]/div[1]/div[3]/div/div/div/div[2]/div/div/div/div[2]/div/div/div/div[2]/div/div[3]/div/div/div/div/div/div[2]/div/div/div/div/div[2]/div/div/div/div/div/button').click()
```

```python

titles = driver.find_elements_by_xpath('//div/h3[@class="css-x53e1b-H3-baseHeadingsCss-baseHeadingsCss-h3Css-h3Css-NameHeader e1fhg9wy5"]/span/span')
categories = driver.find_elements_by_xpath('//div/span/span[@class = "css-8xsu9v-CategoryStyled-CategoryStyled eo1o9660"]')
nofinstalls = driver.find_elements_by_xpath('//div/span[@class="css-f02hti-Installs e13wqmfi8"]')

with open('atlassian data.csv', 'a') as file:
    for i in range(len(titles)):
        file.write(titles[i].text + ";" + categories[i].text + ";" + nofinstalls[i].text + "\n")
```
